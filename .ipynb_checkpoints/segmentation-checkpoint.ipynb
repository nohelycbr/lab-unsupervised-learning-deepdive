```python
#Your code here:

gender = pd.get_dummies(data.gender, drop_first=True)
data_dummy=pd.concat([data, gender], axis=1)
data_dummy.drop('gender',axis=1,inplace=True)
```

```python
data_dummy.head()
```

## 4. Modeling

Now that your data is ready, apply one (or more) unsupervised learning model to find the groups of teens that exist:
* What model did you choose?
* What is the optimal number of clusters?
* Once your clusters are defined, how do you interpret them? - What type of teens there are? What are the differences between them?

```python
#Your code here:

model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,12))
visualizer.fit(data_dummy)
visualizer.poof();
```

```python
model = KMeans(n_clusters = 7)
model.fit(data_dummy)
```

```python
data_dummy['labels'] = model.labels_
```

```python

color = ['royalblue','maroon','forestgreen','mediumorchid','tan','deeppink','olive','goldenrod','lightcyan','navy']
vectorizar = np.vectorize(lambda x: color[x % len(color)])

plt.figure(figsize=(12,7))
plt.scatter(data_dummy.sports, data_dummy.friends, c = vectorizar(data_dummy['labels']))

#centers = model.cluster_centers_
#plt.scatter(centers[:, 0], centers[:, 1], c='red', s=120, alpha=0.75)

plt.show();
```
